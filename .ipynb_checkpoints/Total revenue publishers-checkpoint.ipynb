```python
import re


# Função para converter as abreviações em valores numéricos
def convert_value(value):
    value = str(value)
    value = value.strip()
    if value.startswith('$'):
        value = value[1:]
    # Substituir vírgula por ponto, se presente
    value = value.replace(',', '.')
    # Converter para float
    
    if value.endswith('bn'):
        value = float(value[:-2]) * 1e9  # bilhões
    elif value.endswith('m'):
        value = float(value[:-1]) * 1e6  # milhões
    elif value.endswith('k'):
        value = float(value[:-1]) * 1e3  # milhares
    else:
        if not value:
            value = 0  # Valor vazio, definir como zero
        else:
            #print(value)
            value = float(value)
    return value
```

```python
# Aplicar a função ao DataFrame
df_total_revenue['Total revenue'] = df_total_revenue['Total revenue'].apply(convert_value)
```

```python
 # Verificar se cada valor da coluna 'Publishers' do df_principal está presente no df_total_revenue
for index, row in df_principal.iterrows():
    publishers = row['Publishers']
    total_revenue = 0
    
    # Verificar se o valor de 'publishers' está presente na coluna 'Name' do df_total_revenue
    for name in publishers.split(','):
        mask = df_total_revenue['Name'].str.lower() == name.lower().strip()
        revenue = df_total_revenue.loc[mask, 'Total revenue'].sum()
        total_revenue += revenue
    
    # Inserir o valor total de receita na coluna 'Total revenue' do df_principal
    df_principal.at[index, 'Total revenue'] = total_revenue
```
